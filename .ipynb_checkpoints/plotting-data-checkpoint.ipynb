```python
import pandas as pd
import numpy as np
import sys
import matplotlib.pyplot as plt
from scipy import signal
from scipy.signal import butter, lfilter



def main():
#   input_file = sys.argv[1]
    input_file = 'devin-walk-hand.csv'
    all_data = pd.read_csv(input_file, sep=",")
    total_accel = all_data.drop(['ax', 'ay', 'az'], axis=1)

    # remove the first and last 2 seconds of data 
    total_accel.drop(total_accel.head(200).index, inplace=True)
    total_accel.drop(total_accel.tail(200).index, inplace=True)
    total_accel = total_accel.iloc[::50, :]
    print(total_accel)
    plt.figure(figsize=(30,10))

    plt.plot(total_accel['time'], total_accel['atotal'], 'g-', linewidth=2)
#     plt.savefig('cpu.svg')
    plt.show()

if __name__ == '__main__':
    main()
```

```python
def butter_bandpass(lowcut, highcut, fs, order=5):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return b, a


def butter_bandpass_filter(data, lowcut, highcut, fs, order=5):
    b, a = butter_bandpass(lowcut, highcut, fs, order=order)
    y = lfilter(b, a, data)
    return y


fs = 5000.0
lowcut = 500.0
highcut = 1250.0

y = butter_bandpass_filter(sensor_data, lowcut, highcut, fs, order=6)
plt.plot(sensor_time, y, label='Filtered signal')
plt.xlabel('time (seconds)')
plt.ylabel('Acceleration')
plt.figure(figsize=(100,100))
plt.show()
```
